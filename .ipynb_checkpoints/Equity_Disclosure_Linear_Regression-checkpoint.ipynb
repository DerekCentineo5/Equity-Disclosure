```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as LinMod
import sklearn.metrics as metrics
import sklearn.preprocessing as sp
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
Data = pd.read_csv('Data_4_Equity_Disclosure.csv', header=0)
Data.index = Data['Company']
Data = Data.drop(columns={'Company.1'})
Data
```

```python
X = Data[['Firm_Efficiency']]
y = Data['Reported_EE0-1_Binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
Lin_Regression = LinMod.LinearRegression()
Lin_Regression.fit(X_train, y_train)
```

```python
Lin_Regression.score(X_train, y_train)
```
